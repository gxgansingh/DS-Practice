```python
import pandas as pd
```

1. Exploring Data using Series & Index Objects

```python
# Creating a basic list of marks
marks_list = [85, 90, 88]

# Creating a Series with a custom Index (Labels instead of 0, 1, 2)
# Here, the names are the "Index objects"
student_marks = pd.Series(marks_list, index=['Alice', 'Bob', 'Charlie'])

print(student_marks)
print("\nMarks of Bob:", student_marks['Bob']) # Selecting a specific entry using its custom index
```

2. Exploring Data using Data Frames

```python
# A DataFrame is usually created using a Python Dictionary
# Keys become Column Names, and Lists become the Data (Rows)
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [21, 22, 20],
    'Marks': [85, 90, 88]
}

df = pd.DataFrame(data)
print(df)
```

3. Drop Entry (Removing Data)

```python
# If we want to remove a column, we use the drop() function
# 'axis=1' means we are dropping a COLUMN. ('axis=0' means ROW)
df_without_age = df.drop('Age', axis=1)

print(df_without_age)
```

```python
df_without_bob = df.drop(1, axis=0)

print(df_without_bob)
```

# Example

```python
import pandas as pd
import numpy as np
```

Setup: Creating a basic DataFrame for our operations

```python
data = {
    'Maths': [88, 92, 79, 85],
    'Science': [90, 85, 88, 92]
}
# Using student names as our Index
df = pd.DataFrame(data, index=['Alice', 'Bob', 'Charlie', 'David'])

print("--- Original DataFrame ---")
print(df)
```

1. Selecting Entries (Filtering data)

```python
print("\n--- 1. Selecting Entries ---")
# Using .loc[] to select data by its Index Label
print("Bob's complete data:\n", df.loc['Bob'])

# Selecting specific rows and a specific column
print("\nMaths marks for Alice and Charlie:\n", df.loc[['Alice', 'Charlie'], 'Maths'])
```

2. Re-index (Changing the order or adding new labels)

```python
print("\n--- 2. Re-index ---")
# Changing the order of students and adding a new student 'Eve'
# Since Eve has no original data, Pandas will automatically put NaN (Not a Number)
df_reindexed = df.reindex(['David', 'Charlie', 'Bob', 'Alice', 'Eve'])
print(df_reindexed)
```

3. Rank and Sort

```python
print("\n--- 3. Rank and Sort ---")
# Sorting the table based on Maths marks in descending order (Highest first)
df_sorted = df.sort_values(by='Maths', ascending=False)
print("Students sorted by Maths marks:\n", df_sorted)

# Ranking students in Science (Rank 1 goes to the highest marks)
print("\nRankings in Science:\n", df['Science'].rank(ascending=False))
```

4. Summary Statistics (Quick Math on Data)

```python
print("\n--- 4. Summary Statistics ---")
# .describe() provides count, mean, standard deviation, min, max, etc., all at once!
print("Quick Statistical Summary:\n", df.describe())

# You can also call specific statistical functions like .mean() for average
print("\nAverage marks in each subject:\n", df.mean())
```

5. Data Alignment (Automatic index matching)

```python
print("\n--- 5. Data Alignment ---")
# Creating two separate Series with some overlapping and some different indexes
term1_marks = pd.Series([80, 90], index=['Alice', 'Bob'])
term2_marks = pd.Series([85, 95], index=['Bob', 'Charlie'])

# When we add them, Pandas automatically aligns the data by the Index!
# If an index doesn't exist in BOTH series, the result is NaN.
total_marks = term1_marks + term2_marks
print("Total marks after automatic alignment:\n", total_marks)
```

6. Index Hierarchy (MultiIndex - Multiple levels of indexing)

```python
print("\n--- 6. Index Hierarchy ---")
# Creating a complex index with two levels: 'Class' and 'Student Name'
hierarchical_index = pd.MultiIndex.from_tuples([
    ('Class 10', 'Alice'),
    ('Class 10', 'Bob'),
    ('Class 11', 'Charlie'),
    ('Class 11', 'David')
], names=['Class', 'Student'])

# Applying this hierarchy to a new DataFrame
df_hierarchy = pd.DataFrame(data['Maths'], index=hierarchical_index, columns=['Maths'])
print(df_hierarchy)
```
