```python
import pandas as pd
import numpy as np
```

```python
# Let's create a very messy dataset to clean!
messy_data = pd.DataFrame({
    'Name': ['Alice ', 'BOB', ' Charlie', 'Alice ', 'Eve'], # Messy strings, duplicates
    'Age': [25, np.nan, 35, 25, 999], # Missing data (NaN), Outlier (999)
    'Salary': [50000, 60000, np.nan, 50000, 120000] # Missing data
})

print("--- ORIGINAL MESSY DATA ---\n", messy_data, "\n")
```

1. HANDLING MISSING DATA

```python
print("--- 1. HANDLING MISSING DATA ---")
# Option A: Drop any row that has at least one NaN
df_dropped = messy_data.dropna()
print(df_dropped, "\n")

# Option B: Fill missing Ages with the Mean (Average) Age
# Using inplace=False here just to show the result safely
df_filled = messy_data.copy()
mean_age = df_filled['Age'].mean()
df_filled['Age'] = df_filled['Age'].fillna(mean_age)

print("Filled Missing Ages with Mean:\n", df_filled, "\n")
```

2. DATA TRANSFORMATION & STRING MANIPULATION

```python
print("--- 2. TRANSFORMATION & STRINGS ---")
# Remove exact duplicate rows (Removes the second 'Alice')
df_clean = df_filled.drop_duplicates().copy()

# String Manipulation: Strip extra spaces and convert to Title Case (e.g., 'BOB' -> 'Bob')
df_clean['Name'] = df_clean['Name'].str.strip().str.title()

print("After Removing Duplicates & Cleaning Strings:\n", df_clean, "\n")
```

3. BINNING & CLASSING

```python
print("--- 3. BINNING (Ages to Categories) ---")
# Let's create age groups using pd.cut()
bins = [0, 18, 35, 60, 1000] # Define the edges of the bins
labels = ['Child', 'Young Adult', 'Adult', 'Anomaly'] # Names for the bins

# Apply the bins to the Age column
df_clean['Age_Group'] = pd.cut(df_clean['Age'], bins=bins, labels=labels)
print("Data with Binned Age Groups:\n", df_clean, "\n")
```

4. OUTLIERS & ANOMALIES

```python
print("--- 4. REMOVING OUTLIERS ---")
# We know an age of 999 is impossible (an anomaly/noise).
# Let's filter the DataFrame to only keep rows where Age is less than 100
df_no_outliers = df_clean[df_clean['Age'] < 100].copy()
print("Data after removing age outliers:\n", df_no_outliers, "\n")
```

5. STANDARDIZATION (Z-Score)

```python
# ---------------------------------------------------------
# Standardizing Salary so the numbers aren't massive.
# Formula: (Value - Mean) / Standard Deviation
# ---------------------------------------------------------
print("--- 5. STANDARDIZATION ---")
mean_salary = df_no_outliers['Salary'].mean()
std_salary = df_no_outliers['Salary'].std()

# Standardize the Salary column
df_no_outliers['Standardized_Salary'] = (df_no_outliers['Salary'] - mean_salary) / std_salary

print("Final Cleaned & Standardized Data:\n", df_no_outliers)
```
