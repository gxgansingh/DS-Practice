```python
import pandas as pd
import numpy as np
```

1. MERGING DATASETS (Like SQL Joins or Excel VLOOKUP)

```python
print("--- 1. MERGING DATA ---")
# Create two separate tables (DataFrames)
df_students = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df_scores = pd.DataFrame({'ID': [2, 3, 4], 'Score': [85, 90, 75]})

# Merge them based on the common column 'ID'
# 'inner' merge only keeps IDs that exist in BOTH tables (ID 2 and 3)
merged_inner = pd.merge(df_students, df_scores, on='ID', how='inner')
print("Inner Merge:\n", merged_inner)

# 'outer' merge keeps EVERYTHING. If a match isn't found, it puts NaN (Missing Data)
merged_outer = pd.merge(df_students, df_scores, on='ID', how='outer')
print("\nOuter Merge:\n", merged_outer)
print()
```

2. MERGING ON INDEX

```python
# ---------------------------------------------------------
# Sometimes the common link isn't a column, it's the Index (row labels) itself.
# ---------------------------------------------------------
print("--- 2. MERGE ON INDEX ---")
# Setting 'ID' as the index for both tables
df1 = df_students.set_index('ID')
df2 = df_scores.set_index('ID')

# Merge using left_index and right_index instead of 'on'
merged_on_index = pd.merge(df1, df2, left_index=True, right_index=True, how='inner')
print("Merged on Index:\n", merged_on_index)
print()
```

3. CONCATENATE (Gluing tables together)

```python
print("--- 3. CONCATENATE ---")
df_part1 = pd.DataFrame({'Letter': ['A', 'B']})
df_part2 = pd.DataFrame({'Letter': ['C', 'D']})

# axis=0 glues them top-to-bottom (default)
stacked = pd.concat([df_part1, df_part2], axis=0, ignore_index=True)
print("Top-to-Bottom Concatenation:\n", stacked)
print()
```

4. COMBINING WITH OVERLAP (Patching holes)

```python
print("--- 4. COMBINING WITH OVERLAP ---")
# Table 1 has missing data (NaN)
data1 = pd.Series([10, np.nan, 30, np.nan], index=['a', 'b', 'c', 'd'])
# Table 2 has full data
data2 = pd.Series([100, 200, 300, 400], index=['a', 'b', 'c', 'd'])

# Use data2 to "patch" the missing holes in data1
patched_data = data1.combine_first(data2)
print("Patched Data:\n", patched_data)
print()
```

5. RESHAPING & PIVOTING

```python
print("--- 5. PIVOTING ---")
# Creating a "long" format table (often how databases store data)
sales_data = pd.DataFrame({
    'Date': ['Mon', 'Mon', 'Tue', 'Tue'],
    'City': ['NY', 'LA', 'NY', 'LA'],
    'Sales': [100, 150, 200, 250]
})
print("Original Long Data:\n", sales_data)

# Pivoting turns it into a wide, easy-to-read summary table
# Rows become Dates, Columns become Cities, Values are Sales
pivot_table = sales_data.pivot(index='Date', columns='City', values='Sales')
print("\nPivoted Summary Table:\n", pivot_table)
```
