```python
import matplotlib.pyplot as plt
import numpy as np
```

**1. Import Pyplot**

```python
# Define the coordinates for the x and y axes
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

# Plot the points and display the resulting figure
plt.plot(xpoints, ypoints)
plt.show()
```

**2. 1D Array of Subplots**

```python
# plot 1: Create a data series and place it in a 1x2 grid at position 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1) # Parameters are: (rows, columns, index)
plt.plot(x,y)

# plot 2: Create another data series and place it in the same grid at position 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(x,y)

# Render both subplots side-by-side
plt.show()
```

**3. 1D Array of Subplots**

```python
# plot 1: Create a data series and place it in a 2x1 grid at position 1 (top)
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 1, 1)
plt.plot(x,y)

# plot 2: Create another data series and place it in the same grid at position 2 (bottom)
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 1, 2) # Parameters are: (rows, columns, index)
plt.plot(x,y)

# Display the vertical stack of plots
plt.show()
```

**4. 2D Array of Subplots**

```python
# Populate a 2x3 grid with different data points in each subplot position
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# Subplot index 1 to 6
plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array( [10, 20, 30, 40])
plt.subplot(2, 3, 6)
plt.plot(x,y)

# Render the full 2x3 grid
plt.show()
```

**5. Multiple plots**

```python
# Create a figure and a 2x2 grid of subplots (axes object)
fig, axs = plt.subplots(2, 2)

# Plot a line graph in the top-left subplot
axs[0, 0].plot([1, 2, 3], [4, 5, 6])

# Plot a scatter graph in the top-right subplot
axs[0, 1].scatter([1, 2, 3], [4, 5, 6])

# Plot a bar chart in the bottom-left subplot
axs[1, 0].bar([1, 2, 3], [4, 5, 6])

# Plot a histogram in the bottom-right subplot
axs[1, 1].hist([1, 2, 2, 3, 3, 3, 3, 3, 3, 4, 4, 5])

# Adjust and show the figure
plt.show()
```

# Controlling axes, ticks, labels, legends

**Controlling axes**

```python
# Create a simple line plot
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Set the specific limits for the X and Y axes
plt.xlim(0, 5)
plt.ylim(0, 35)

# Display the plot with adjusted boundaries
plt.show()
```

**Controlling Ticks**

```python
# Plot simple line data
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Manually set the positions and text labels for the X-axis
plt.xticks([1, 2, 3, 4], ['One', 'Two', 'Three', 'Four'])

# Manually set the positions and text labels for the Y-axis
plt.yticks([10, 20, 30], ['Low', 'Medium', 'High'])

# Display the plot with customized ticks
plt.show()
```

```python
# Plot data points
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Rotate the tick labels to prevent overlap or for style
plt.xticks(rotation=45) # Rotate x-axis ticks by 45 degrees
plt.yticks(rotation=90) # Rotate y-axis ticks by 90 degrees

# Show the plot with rotated labels
plt.show()
```

**Controlling Labels**

```python
#Setting Axis Labels
#Axis labels can be set using xlabel() and ylabel()
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.show()
```

```python
# Plot data points
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Set labels with custom font size and colors
plt.xlabel('X Axis', fontsize=14, color='blue')

# Set Y label with custom properties
plt.ylabel('Y Axis', fontsize=14, color='red')

# Show the customized label plot
plt.show()
```

```python
# Plot data points
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Set a central title for the entire plot with a larger font size
plt.title('Plot Title', fontsize=16)

# Render the plot
plt.show()
```

**Controlling Legends**

```python
# Plot two different lines and assign labels for the legend
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], label='Line 1')
plt.plot([1, 2, 3, 4], [30, 25, 20, 15], label='Line 2')

# Display the legend based on the 'label' parameters provided above
plt.legend()

# Show the plot
plt.show()
```

```python
# Plot two lines with labels
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], label='Line 1')
plt.plot([1, 2, 3, 4], [30, 25, 20, 15], label='Line 2')

# Position the legend explicitly in the upper left corner
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

# Annotations, saving plots

• **Text annotations** are used to add explanatory or descriptive text to specific points, regions, or features within a plot.

• **Marker annotations** involve placing markers or symbols on specific points of interest within a plot to highlight or provide additional information about those points.

• **Callouts** refer to a specific type of annotation that uses visual elements like arrows, lines, or text to draw attention to a particular area or feature within a plot.

**Simple Text Annotation**

```python
# Prepare data coordinates
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a plot with markers at data points
plt.plot(x, y, marker='o')

# Add text at a specific coordinate (xytext) without an arrow pointing to the data (xy)
plt.annotate('Simple Text Annotation',
             xy=(4, 8),
             xytext=(2.5, 9),
             fontsize=12, color='red')

# Display the plot with the floating text
plt.show()
```

**Annotation with Arrow**

```python
# Prepare data and plot a styled line with markers
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y, marker='o', linestyle='-', color='blue')

# Add an annotation with an arrow pointing to a specific data point
plt.annotate('Peak', xy=(3, 5), xytext=(3.5, 6),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=10)

# Set labels, title, and enable grid for better readability
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Annotated Plot')
plt.grid(True)
plt.show()
```

**Annotation with different Arrow Styles**

```python
# Define coordinates and plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, marker='o')

# Annotate with a purple bracket-style arrow
plt.annotate('Fancy Arrow',
             xy=(2, 4), xytext=(1.5, 5),
             arrowprops=dict(arrowstyle='-[', facecolor='purple'))

# Annotate with an orange bar-ended arrow
plt.annotate('Bar Arrow',
             xy=(5, 10), xytext=(4.5, 11),
             arrowprops=dict(arrowstyle='|-|', facecolor='orange'))

# Display the variations
plt.show()
```

**Annotation with bounding box around the text**

```python
# Define coordinates and plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, marker='o')

# Annotate with a text box (bbox) and an arrow
plt.annotate('With Bounding Box', xy=(3, 6), xytext=(4, 7),
             bbox=dict(boxstyle='round,pad=0.5', edgecolor='black', facecolor='yellow'),
             arrowprops=dict(facecolor='blue', arrowstyle='->'))

# Render the stylized annotation
plt.show()
```

**Saving Plots**

```python
# Prepare line plot data
x = [22,1,7,2,21,11,14,5]
y = [24,2,12,5,5,5,9,12]

# Generate the line plot
plt.plot(x,y)

# Set labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Export the plot to a PNG file in the current directory
plt.savefig('my_plot.png')

# Render the plot to the screen
plt.show()
```
