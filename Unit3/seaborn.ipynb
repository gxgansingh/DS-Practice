```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from mpl_toolkits.mplot3d import Axes3D
```

**Scatter Plot**

```python
data = {
 "x": [1, 2, 3, 4, 5],
 "y": [2, 4, 3, 5, 1],
 "hue": ["A", "B", "A", "B", "A"] }
df = pd.DataFrame(data)
# Create a scatter plot with hue to differentiate categories
sns.scatterplot(x="x", y="y", hue="hue", data=df)
# Add a title and labels
plt.title("Scatter Plot with Categories")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Load the tips dataset
tips = sns.load_dataset("tips")
# Create a scatter plot
sns.scatterplot(x="total_bill", y="tip", data=tips)
# Add a title and labels
plt.title("Total Bill vs. Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
# Show the plot
plt.show()
```

**Joint Plots**

```python
data = {
 "x": [1, 2, 3, 4, 5],
 "y": [2, 5, 3, 8, 10],
 }
# Create a joint plot which shows both the bivariate relationship and univariate distributions
sns.jointplot(x='x', y='y', kind="scatter", data=data)
plt.show()
```

**Pair Plots**

```python
data = {
 "x": [1, 2, 3, 4, 5],
 "y": [2, 4, 3, 5, 1],
 "z": [10, 15, 20, 25, 30]
}
df = pd.DataFrame(data)
# Generate a matrix of plots showing pairwise relationships in a dataset
sns.pairplot(df)
plt.show()
```

**Grid Plot**

```python
# Initialize a grid with 2 rows and 3 columns
Grid_plot = plt.GridSpec(2, 3, wspace = 0.8, hspace = 0.6)

# Create a subplot at the first row, first column
plt.subplot(Grid_plot[0, 0])
plt.grid(True)

# Create a subplot spanning the first row, second and third columns
plt.subplot(Grid_plot[0, 1:])
plt.grid(True)

# Create a subplot spanning the second row, first and second columns
plt.subplot(Grid_plot[1, :2])
plt.grid(True)

# Create a subplot at the second row, third column
plt.subplot(Grid_plot[1, 2])
plt.grid(True)
```

**Sub Plots**

# Multiple Plots

**Facet Grid**

```python
data = {
 "x": [1, 2, 3, 4, 5],
 "y": [2, 4, 3, 5, 1],
 "hue": ["A", "B", "A", "B", "A"]
}
df = pd.DataFrame(data)
# Create a FacetGrid with one row and two columns
g = sns.FacetGrid(df, col="hue", height=4)
# Plot scatter plots for each facet
g.map(sns.scatterplot, "x", "y")
plt.show()
```

# Scatterplot, Lineplot and Histogram

**Scatterplot**

```python
data = {
 "x_values": [1, 2, 3, 4, 5],
 "y_values": [2, 5, 3, 8, 10] }
# Create a scatterplot
sns.scatterplot(x="x_values", y="y_values", data=data)
# Add a title and labels
plt.title("Scatterplot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# Show the plot
plt.show()
```

**Line Plot**

```python
x_values = [1, 2, 3, 4, 5]
y_values = [2, 5, 3, 8, 10]
# Create the line plot
sns.lineplot(x=x_values, y=y_values)
# Add a title and labels
plt.title("Sample Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# Show the plot
plt.show()
```

**Histogram**

```python
# Generate 1000 random points from a normal distribution
data = np.random.randn(1000)
# Create a histogram to show the density distribution of the data
plt.hist(data, density=True)
# Label the plot
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

# Boxplot

```python
data = {
 "A": [1, 2, 3, 4, 5],
 "B": [2, 4, 6, 8, 10],
 "C": [3, 6, 9, 12, 15] }
# Create a box plot
sns.boxplot(data=data)
# Add a title and labels
plt.title("Box Plot Example")
plt.xlabel("Columns")
plt.ylabel("Values")
# Show the plot
plt.show()
```

```python
data = {
 "A": [1, 2, 3, 4, 5, 30], # 30 is an outlier
 "B": [2, 4, 6, 8, 7, 28], # 28 is an outlier
 "C": [3, 6, 9, 5, 2, 7]
}
# Convert data to DataFrame for better visualization
df = pd.DataFrame(data)
# Create a box plot with outliers explicitly shown
sns.boxplot(data=df, showmeans=True, whis=1.5)
# Add a title and labels
plt.title("Box Plot with Outliers")
plt.xlabel("Columns")
plt.ylabel("Values")
# Show the plot
plt.show()
```

```python
data = {
 "A": [10, 20, 30, 40, 50],
 "B": [5, 10, 15, 20, 25],
 "C": [8, 12, 16, 20, 24],
 "D": ["A", "B", "A", "B", "A"] }
# Convert data to DataFrame
df = pd.DataFrame(data)
# Create a pair plot
sns.pairplot(df, hue="D")
# Use 'D' for color differentiation
plt.show()
```

**Word Cloud**

```python
# Sample text data
text = "This is a sample text for creating a word cloud. Word clouds are a great way to visualize the frequency of words in a text."
# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400,
background_color="white")
# Generate the word cloud
wordcloud.generate(text)
# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

**Text Annotation on a Plot**

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 3, 5, 1]
# Create a scatter plot
sns.scatterplot(x=x, y=y)
# Add text annotations
for i in range(len(x)):
 plt.text(x[i], y[i], f"({x[i]}, {y[i]})",
ha="center", va="bottom")
plt.show()
```

**Text as a Plot Element**

```python
fig, ax = plt.subplots()
# Add text to the plot
ax.text(0.5, 0.5, "This is a sample text on the plot",
 horizontalalignment='center',
 verticalalignment='center',
 transform=ax.transAxes)
plt.show()
```

# **3D PLOT**

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 3, 5, 1]
z = [10, 15, 20, 25, 30]
# Create a 3D figure object
fig = plt.figure()
# Add a 3D subplot
ax = fig.add_subplot(111, projection='3d')
# Plot a 3D scatter plot
scatter = ax.scatter(x, y, z, c='blue', s=100)
# Set labels for the three dimensions
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('3D Scatter Plot')
# Set the viewing angle
ax.view_init(elev=20, azim=45)
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 3, 5, 1]
z = [4, 5, 3, 2, 6]
plt.figure(figsize=(6,5))
plt_axis = plt.axes (projection = '3d')
plt_axis.plot3D (x, y, z)
plt.tight_layout ()
plt.title('3D Line Plot')
plt.show ()
```

```python
# Sample data (ensure x and y are 2D for surface plot)
x = np.linspace(0, 5, 10) # Create equally spaced points from 0 to 5 with 10 elements
y = np.linspace(0, 5, 10)
X, Y = np.meshgrid(x, y) # Create a 2D grid from x and y for surface evaluation
# Define a function to evaluate z values at each grid point
def f(x, y):
 return x**2 + y**2 # Replace with your desired function
# Calculate z values based on the function
z = f(X, Y)
# Create a 3D figure and axes
fig = plt.figure(figsize=(8, 6)) # Adjust figure size as needed
ax = fig.add_subplot(111, projection='3d')
# Create a surface plot
surf = ax.plot_surface(X, Y, z, cmap='viridis', linewidth=0, antialiased=True) # Adjust colormap
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
plt.title('3D Surface Plot')
# Customize viewing angle (optional)
ax.view_init(elev=20, azim=45) # Adjust elevation and azimuth angles
# Show the plot
plt.show()
```
